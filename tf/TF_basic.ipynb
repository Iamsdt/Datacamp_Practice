```python
import tensorflow as tf

# define variable

a = tf.Variable([1, 2, 3, 4], dtype=tf.float32)
a
```

```python
# define constant
b = tf.constant([1, 2, 4, 5], dtype=tf.float32)
b
```

```python
tf.multiply(a, b).numpy()
```

```python
(a * b).numpy()
```

```python
from tensorflow import ones, matmul, multiply

A0 = ones(1)
A31 = ones([3, 1])
A34 = ones([3, 4])
A43 = ones([4, 3])
```

```python
print(multiply(A0, A0))
print(multiply(A31, A31))
print(multiply(A34, A34))
```

```python
matmul(A43, A34 )
```

```python
matmul(A43, A43)
```

```python
from tensorflow import ones, reduce_sum
# Define a 2x3x4 tensor of ones
A = ones([2, 3, 4])
```

```python
# Sum over all dimensions
B = reduce_sum(A)
B
```

```python
# Sum over dimensions 0, 1, and 2
B0 = reduce_sum(A, 0)
B1 = reduce_sum(A, 1)
B2 = reduce_sum(A, 2)
```

```python
B0, B1, B2
```

```python
reduce_sum(A31, 1)
```

```python
x = tf.Variable(-1.0)
```

```python
with tf.GradientTape() as tape:
    tape.watch(x)
    y = tf.multiply(x, x)
```

```python
g = tape.gradient(y, x)
```

```python
g.numpy()
```
